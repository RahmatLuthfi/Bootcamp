```python
#Import Library Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import Library Numpy
import numpy as np

#Menghitung jumlah data 100 yang berjarak dari 0 sampai 2Ï€
x = np.linspace(2,0*np.pi, 100)
#Kemudian menghitung cos si x
cos_x = np.cos(x)
```

```python
#Membuat line pilot
fig, ax = plt.subplots()
_ = ax.plot (x, cos_x)
```

```python
#Membuat sumbu x dan y memiliki rasio yang sama
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x, markersize=20, linestyle='-.',
            color='red', label='cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```
