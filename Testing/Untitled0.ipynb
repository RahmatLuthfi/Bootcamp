```python
#Import Library Matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import Library Numpy
import numpy as np

#Menghitung jumlah data 100 yang berjarak dari 0 sampai 2Ï€
x = np.linspace(0,9*np.pi, 100)
#Kemudian menghitung cos si x
cos_x = np.cos(x)
```

```python
#Membuat line pilot
fig, ax = plt.subplots()
_ = ax.plot (x, cos_x)
```

```python
#Membuat sumbu x dan y memiliki rasio yang sama
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
_ = ax.set_aspect('equal')
```

```python
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x, markersize=20, linestyle='-.',
            color='red', label='cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

```python
#Membuat scatter plot
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

fig, ax = plt.subplots()
#Membuat scatter plot
_ = ax.scatter(x, y)
#set label pada variabel x dan y
_ = ax.set_xlabel('x axis')
_ = ax.set_ylabel('y axis')
```
